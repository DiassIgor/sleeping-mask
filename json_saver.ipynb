```python
import pandas as pd
import requests
import functions
```

```python
def get_data(route, start, end):
    
    headers = {'Authorization': 'Bearer CZQY7GXLYM4JRPEWO4SYGLCRFG5BXD5K'}
    start_date = start
    end_date = end
    url = 'https://api.ouraring.com/v2/usercollection/'
    
    
    if route == 'heart':
        
        sufix = 'T00:00:00-00:00'
        params={'start_datetime': start_date+sufix, 'end_datetime': end_date+sufix}
        
    else:
        params={'start_date': start_date, 'end_date': end_date}
    
    response = requests.request('GET', url+route, headers=headers, params=params) 
    df = pd.DataFrame(response.json()['data'])
    
    return df
```
